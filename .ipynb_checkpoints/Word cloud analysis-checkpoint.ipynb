```python
import tweepy
from os import path
from scipy.misc import imread
import matplotlib.pyplot as plt
import random
# import the wordcloud for ploting the cloud of words
from wordcloud import WordCloud, STOPWORDS # STOPWORDS is a set
```

```python
# load data
file = open('/Users/dingyunxing/Documents/AnacondaProjects/tweetTrump/tweet/all_tweets.txt', 'r')
tweets = file.read()
file.close()
```

### Plot the cloud

```python
wordcloud = WordCloud(#font_path='/Library/Fonts/Verdana.ttf',
                      relative_scaling = 0.8,
                      width=1600, height=800,
                      stopwords = STOPWORDS.update(['https','co','amp']) # set or space-separated string
                      ).generate(tweets)
plt.figure(figsize=(18,10), dpi=500)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
plt.savefig('/Users/dingyunxing/Documents/AnacondaProjects/tweetTrump/plot/cloudword.png', dpi=500, format='png')
```
