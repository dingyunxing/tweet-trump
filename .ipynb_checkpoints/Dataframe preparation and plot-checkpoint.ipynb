```python
import pandas as pd     
import numpy as np      
import matplotlib.pyplot as plt
%matplotlib inline
```

### read and clean dataset

```python
df = pd.read_csv("/Users/dingyunxing/Documents/AnacondaProjects/tweetTrump/tweet/df.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
# split the date column into year, month, day...
time = pd.DataFrame(df.Date.str.split(' ').tolist(), columns = ['Date','Mon','Day','Time','Zone','Year'])
```

```python
# combine the datasets
df = pd.concat([df, time], axis=1)
```

```python
# drop the orginal date column
df = df.drop(['Date'], axis=1)
```

```python
# change the month of abbreviation to numbers
import calendar
list(calendar.month_abbr).index("Jan")
for mon in df['Mon']:
    if isinstance(mon, str):
        trans = list(calendar.month_abbr).index(mon)
        df.replace(mon, trans, inplace=True)
```

```python
# have a look at the clean dataset
df.head()
```

```python
# see the sum of tweets for every month
month = df.groupby(['Year', 'Mon']).size()

# see the sum of likes and retweets for each month
combine = df.groupby(['Year', 'Mon'])['RTweets','Likes'].agg('sum')
```

```python
month
```

```python
combine
```

### Plot the result

```python
plt.figure(figsize=(18,10))
plt.title("The number of Trump's tweets every month",fontsize=20)
plt.xlabel("numbers")
month.plot(kind="bar")
plt.savefig('/Users/dingunxing/Documents/AnacondaProjects/tweetTrump/plot/numbers by month.png', dpi=300, format='png')
```

```python
pl = combine.plot(kind='bar', title ="The compare of retweets and likes of each month",figsize=(15,10),legend=True, fontsize=20)
pl.set_ylabel("numbers(million)",fontsize=12)
```
