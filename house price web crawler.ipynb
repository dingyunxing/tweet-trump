```python
import time
import pandas as pd
import numpy as np

from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.Chrome()

url = 'https://www.trademe.co.nz/property/insights/address/Christchurch/Ilam/Hamilton-Avenue/29C/691d17f0-5da2-46d5-b90a-8e375e88f5ea'

browser.get(url)
#time.sleep(1)
contents = []
while len(url) > 0:
    browser.get(url)
    body = browser.find_element_by_tag_name('body')

#for _ in range(5):
 #   body.send_keys(Keys.PAGE_DOWN)
  #  time.sleep(0.05)
    
    tables = browser.find_elements_by_class_name('property-details')
#times = browser.find_elements_by_class_name('_timestamp')
    prices = browser.find_elements_by_class_name('timeline-container')
    links = browser.find_elements_by_xpath("//a[@href]")
    for link in links:
        if len(link.get_attribute("href")) > 120:
            url = link.get_attribute("href")
    for table in tables:
        contents.append(table.text)
```
